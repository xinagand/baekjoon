```python
# 리스트, 셋, 딕셔너리, deque 속도정리

from collections import deque

# 리스트(List) 사용 예시
lst = [1, 2, 3, 4, 5]
lst.append(6)           # 리스트 끝에 추가 (O(1))
lst.pop(0)              # 리스트 앞에서 제거 (O(n)): 중간이나 앞에서의 삽입/삭제 비효율적

# 셋(Set) 사용 예시
s = {1, 2, 3, 4, 5}
s.add(6)                # 새로운 요소 추가 (O(1)), 중복은 자동 제거
s.remove(3)             # 특정 요소 제거 (O(1)), 데이터 순서 보장 안 됨

# 딕셔너리(Dictionary) 사용 예시
d = {"a": 1, "b": 2, "c": 3}
d["d"] = 4              # 키-값 쌍 추가 (O(1))
del d["b"]              # 특정 키 삭제 (O(1)), 메모리 사용량 높음

# 디큐(Deque) 사용 예시 ~ 인덱스로 참조할 땐 느려짐(링크드리스트)
dq = deque([1, 2, 3, 4, 5])
dq.append(6)            # deque 끝에 추가 (O(1))
dq.appendleft(0)        # deque 앞에 추가 (O(1))
dq.pop()                # deque 끝에서 제거 (O(1))
dq.popleft()            # deque 앞에서 제거 (O(1)), 중간 요소 접근은 불편 (O(n))
```

```python
# 입력 받는 방법의 차이
import sys
import time
import io

# 비교할 데이터의 크기 설정 (테스트 시 입력 데이터를 준비해야 합니다)
input_data = "asodinqowid qsoidnqowind qsoidnoq\n".join(map(str, range(10000)))  # 예시로 10만 줄의 데이터 생성
sys.stdin = io.StringIO(input_data)

# input()을 이용한 실행 시간 측정 (여기서는 시뮬레이션을 위해 splitlines 사용)
start_time = time.time()
for line in range(100000):  # line 단위로 input()을 시뮬레이션
    line = input()  # 실질적인 작업을 넣을 수 있습니다.
input_time = time.time() - start_time

# sys.stdin.read()를 이용한 실행 시간 측정
sys.stdin = io.StringIO(input_data)
start_time = time.time()
data = sys.stdin.read()
lines = data.splitlines()  # 한 번에 모든 줄을 읽고 나서 분리
for line in lines:
    line = line
read_time = time.time() - start_time

# sys.stdin.readlines()를 이용한 실행 시간 측정
sys.stdin = io.StringIO(input_data)
start_time = time.time()
lines = sys.stdin.readlines()
for line in lines:
    line = line.strip()
readlines_time = time.time() - start_time

# 결과 출력
print(f"input() method: {input_time:.5f} seconds")
print(f"sys.stdin.read() method: {read_time:.5f} seconds")
print(f"sys.stdin.readlines() method: {readlines_time:.5f} seconds")
```
